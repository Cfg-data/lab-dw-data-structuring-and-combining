```python
import numpy as np
import pandas as pd

insurance_data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')

insurance_data.reset_index(drop=True, inplace=True)
insurance_data.columns = [c.lower().strip() for c in insurance_data.columns.values.tolist()]

# Assuming insurance_data is defined and contains the 'st' column
insurance_data.rename(columns={'st': 'state'}, inplace=True)

unique_states = insurance_data['state'].unique()

unique_states = insurance_data['state'].replace('WA', 'Washington', inplace=True)
unique_states = insurance_data['state'].replace('AZ', 'Arizona', inplace=True)
unique_states = insurance_data['state'].replace('Cali', 'California', inplace=True)

unique_states = insurance_data['gender'].unique()

unique_states = insurance_data['gender'].replace('Femal', 'F', inplace=True)
unique_states = insurance_data['gender'].replace('female', 'F', inplace=True)
unique_states = insurance_data['gender'].replace('Male', 'M', inplace=True)

unique_states = insurance_data['education'].unique()

unique_states = insurance_data['education'].replace('Bachelors', 'Bachelor', inplace=True)

unique_states = insurance_data['customer lifetime value'].unique()

insurance_data['customer lifetime value'] = insurance_data['customer lifetime value'].str.replace('%', '').astype(float)

unique_states = insurance_data['vehicle class'].unique()

unique_states = insurance_data['vehicle class'].replace('Sports Car', 'Luxury', inplace=True)
unique_states = insurance_data['vehicle class'].replace('Sports Car', 'Luxury', inplace=True)
unique_states = insurance_data['vehicle class'].replace('Luxury Car', 'Luxury', inplace=True)

unique_states = insurance_data['customer lifetime value'].unique()

unique_states = insurance_data['income'].unique()

unique_states = insurance_data['number of open complaints'].unique()

unique_states = insurance_data['number of open complaints'].replace('1/0/00', '0', inplace=True)
unique_states = insurance_data['number of open complaints'].replace('1/2/00', '2', inplace=True)
unique_states = insurance_data['number of open complaints'].replace('1/1/00', '1', inplace=True)
unique_states = insurance_data['number of open complaints'].replace('1/3/00', '3', inplace=True)
unique_states = insurance_data['number of open complaints'].replace('1/5/00', '5', inplace=True)
unique_states = insurance_data['number of open complaints'].replace('1/4/00', '4', inplace=True)

insurance_data.dropna(inplace=True)
print(insurance_data.shape)

# Convert all numeric columns to int
insurance_data.select_dtypes(include=['float64', 'int64']).astype(int)

# Keep changes in DataFrame
insurance_data[insurance_data.select_dtypes(include=['float64']).columns] = insurance_data.select_dtypes(include=['float64']).astype(int)

print(insurance_data.dtypes)

# Check for duplicate rows
duplicates = insurance_data.duplicated()

# Display the boolean Series
print(duplicates)

# Optionally, you can filter and show the duplicate rows
duplicate_rows = insurance_data[duplicates]
print(duplicate_rows)

# If you want to drop duplicates:
insurance_data.drop_duplicates(inplace=True)

# Reset the index
insurance_data.reset_index(drop=True, inplace=True)

# Display the updated DataFrame
print(insurance_data)

insurance_data.head()
```
